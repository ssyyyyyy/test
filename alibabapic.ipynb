```python
import pandas as pd  # 导入pandas库
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
# 修正后的代码

# 示例数据，您应该用您的实际数据来替换这些
# 这里我们随机生成一些数据来模拟您的数据结构
np.random.seed(10)
data = {
    '0.2-0.4': np.random.normal(loc=20, scale=5, size=100),
    '1': np.random.normal(loc=30, scale=5, size=100),
    '3': np.random.normal(loc=40, scale=5, size=100),
    '6': np.random.normal(loc=50, scale=5, size=100),
    '12': np.random.normal(loc=60, scale=5, size=100),
    '24': np.random.normal(loc=70, scale=5, size=100),
    '48': np.random.normal(loc=80, scale=5, size=100)
}

# 使用pandas创建DataFrame
df = pd.DataFrame(data)

# 创建箱形图
plt.figure(figsize=(12, 6))
sns.boxplot(data=df)

# 设置图表标题和轴标签
plt.title('Boxplot of Reclaimable Idle Memory')
plt.xlabel('Age Range of Pages (Hour)')
plt.ylabel('Percentage (%)')

# 显示图形
plt.show()
plt.savefig('C:\pythonProject\littlecode\alibabapic\xiangxiantu.png')

```
